```python
import tensorflow as tf
from tensorflow import keras
```

```python
tf.__version__, keras.__version__
```

```python
#IMPORT DATA
fashion_mnist=keras.datasets.fashion_mnist
(X_train_full, y_train_full), (X_test, y_test)=fashion_mnist.load_data()
```

```python
X_train_full.shape, X_train_full.dtype
```

```python
X_valid, X_train=X_train_full[:5000]/255.0, X_train_full[5000:]/255.0
y_valid, y_train=y_train_full[:5000], y_train_full[5000:]
X_test = X_test / 255.
```

```python
class_names = ["T-shirt/top", "Trouser", "Pullover", "Dress", "Coat",
                   "Sandal", "Shirt", "Sneaker", "Bag", "Ankle boot"]
```

```python
class_names[0]
```

```python
y_train[0:2].astype('int')
```

```python
y_train.shape,  y_train.shape
```

```python
plt.imshow(X_train[0], cmap="binary")
plt.axis('off')
plt.show()
```

```python
n_rows = 5
n_cols = 20
plt.figure(figsize=(n_cols * 1.2, n_rows * 1.2))
for row in range(n_rows):
    for col in range(n_cols):
        index = n_cols * row + col
        plt.subplot(n_rows, n_cols, index + 1)
        plt.imshow(X_train[index], cmap="binary", interpolation="nearest")
        plt.axis('off')
        plt.title(class_names[y_train[index]], fontsize=12)
plt.subplots_adjust(wspace=0.2, hspace=0.5)
save_fig('fashion_mnist_plot', tight_layout=False)
plt.show()
```

```python
X_train[1].shape
```

```python
model=keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[28,28]))
model.add(keras.layers.Dense(300, activation='relu'))
model.add(keras.layers.Dense(100, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
model.layers
```

```python
model.layers[1].name
```

```python
import pydot
import graphviz
```

```python
keras.utils.plot_model(model, "my_fashion_mnist_model.png", show_shapes=True)
```

```python
hidden1 = model.layers[1]
hidden1.name
```

```python
weights, biases=hidden1.get_weights()
```

```python
#NOTICE HOW WEIGHTS ARE ALL RANDOM AND BIASES ARE ZERO
weights, biases
```

```python
biases.shape
```

```python
#Compiling the model
model.compile(loss='sparse_categorical_crossentropy',
             optimizer='sgd',
             metrics=['accuracy'])
```

```python
history=model.fit(X_train, y_train, epochs=30,
                 validation_data=(X_valid, y_valid))
```

```python
import pandas as pd
```

```python
pd.DataFrame(history.history).plot(figsize=(10, 6))
plt.grid(True)
plt.gca().set_ylim(0,1)
```

```python
model.evaluate(X_test, y_test)
```

```python
#MAKE PREDICTIONS
X_new=X_test[:100]
y_prob=model.predict(X_new)
y_prob.round(2)
```

```python
#y_pred = model.predict_classes(X_new)

y_pred = np.argmax(model.predict(X_new), axis=-1)
y_pred
```

```python
Unmatched=y_test[:100]==y_pred
Unmatched_Index=np.where(Unmatched==False)

Unmatched_X=X_new[Unmatched==False]
Unmatched_y=y_pred[Unmatched==False]

Correct_y=y_test[:100][Unmatched==False]
```
